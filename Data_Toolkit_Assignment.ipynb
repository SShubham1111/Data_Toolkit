# Theoritical Questions

1. What is NumPy, and why is it widely used in Python
- NumPy is a powerful Python library offering efficient array operations, essential for numerical computing, data manipulation, and scientific analysis.
- It provides high-performance multidimensional arrays and mathematical functions, enabling optimized linear algebra, Fourier transforms, and random number generation.
- NumPy integrates seamlessly with libraries like SciPy, Pandas, and scikit-learn, making it foundational for data science workflows.

2. How does broadcasting work in NumPy
- Broadcasting allows operations on arrays of different shapes by automatically expanding smaller arrays without explicit replication.
- It simplifies element-wise operations across mismatched dimensions, improving readability, speed, and reducing manual loops.
- Broadcasting ensures computational efficiency and memory optimization when handling arithmetic operations across multidimensional arrays.

3. What is a Pandas DataFrame
- DataFrame is a two-dimensional labeled data structure in Pandas, similar to a relational database table.
- It stores data in rows and columns, supporting heterogeneous data types for versatile analysis and manipulation tasks.

4. Explain the use of the groupby() method in Pandas
- groupby() groups data based on specified keys, enabling aggregation, transformation, or filtering of subsets efficiently.
- supports applying functions like sum, mean, or count on groups, providing concise and meaningful data summaries.
- This method is crucial for split-apply-combine operations, often used in exploratory analysis and reporting tasks.

5. Why is Seaborn preferred for statistical visualizations
- Seaborn provides high-level APIs that simplify statistical visualizations, producing aesthetically pleasing charts with minimal code.
- It integrates tightly with Pandas DataFrames, automatically handling categorical variables and missing values gracefully.
- Seaborn supports advanced statistical plots like violin, swarm, and regression plots, improving interpretability in data exploration.

6. What are the differences between NumPy arrays and Python lists
- NumPy arrays are homogeneous, requiring consistent data types, while Python lists can store heterogeneous elements.
- Arrays support vectorized operations and broadcasting, making numerical computations efficient compared to slower element-wise Python list operations.
- NumPy arrays use contiguous memory blocks, improving performance, while Python lists store references, causing overhead.

7. What is a heatmap, and when should it be used
- heatmap is a graphical representation of data using color coding to show magnitude in a matrix format.
- Used for correlation matrices, confusion matrices, and feature importance visualization in analytics tasks.
- Heatmaps provide quick insights into variable relationships, patterns, and anomalies, supporting effective exploratory data analysis.

8. What does the term “vectorized operation” mean in NumPy
- Vectorized operations apply computations directly to arrays without explicit loops, leveraging optimized C-based implementations for speed.They simplify code, enhance readability, and eliminate manual iteration when performing arithmetic or logical operations.
- vectorization significantly improves performance by minimizing overhead, enabling efficient large-scale data processing

9. How does Matplotlib differ from Plotly
- Matplotlib is static, producing publication-quality 2D plots, whereas Plotly specializes in interactive, dynamic visualizations.
- Plotly supports built-in zooming, panning, and tooltips, enhancing exploratory analysis, unlike Matplotlib static images and integrates deeply with scientific workflows, while Plotly is preferred for dashboards, web applications, and user interactivity.

10. What is the significance of hierarchical indexing in Pandas
- Hierarchical indexing enables multiple levels of row or column labels, improving data representation in complex datasets.
- Simplifies subsetting and reshaping, supporting advanced operations like pivoting and slicing multidimensional structures.
- Hierarchical indexing enhances readability and flexibility in multi-key groupings and time series analysis.

11. What is the role of Seaborn’s pairplot() function
- pairplot() creates scatterplot matrices for multiple variables, visualizing pairwise relationships and distributions efficiently.
- Helps detect correlations, trends, and outliers by showing multiple variable combinations simultaneously and is often used in exploratory analysis for multivariate datasets.

12. What is the purpose of the describe() function in Pandas
- describe() generates summary statistics of numerical data, including mean, standard deviation, minimum, and percentiles.
- It helps quickly understand central tendency, spread, and overall data distribution without manual calculations.
- The function supports numeric, categorical, and datetime data, providing tailored summaries for different variable types.

13. Why is handling missing data important in Pandas
- Missing data can bias analysis results, leading to incorrect conclusions if not addressed properly.
- Handling ensures dataset integrity, supporting accurate statistical modeling, predictive analytics, and machine learning outcomes.
- Pandas offers tools like fillna(), dropna(), and interpolation for flexible missing data management.

14. What are the benefits of using Plotly for data visualization
- Plotly supports interactive plots with tooltips, zooming, and filtering, improving exploratory and presentation experiences.
- Integrated with Python, R, and JavaScript, enabling cross-platform dashboard development.
- Plotly supports advanced 3D visualizations, geospatial mapping, and streaming data visualization seamlessly.

15. How does NumPy handle multidimensional arrays
- NumPy provides the ndarray object, supporting N-dimensional arrays with uniform data types and efficient memory layouts.
- It offers powerful indexing, slicing, reshaping, and broadcasting capabilities for multidimensional data structures.
- NumPy uses strides and shape metadata to access elements efficiently without replicating data in memory.

16. What is the role of Bokeh in data visualization
- Bokeh creates interactive, browser-based visualizations that support zooming, panning, and linked selections.
- It integrates with Python and web technologies, enabling dynamic dashboards and streaming data apps.
- Bokeh is ideal for real-time visual analytics in web applications.

17. Explain the difference between apply() and map() in Pandas
- map() applies functions element-wise on Series, suitable for simple transformations.
- apply() applies functions on Series or DataFrames, supporting row-wise and column-wise operations, is more versatile, while map() is efficient for element-level transformations.

18. What are some advanced features of NumPy
- NumPy supports broadcasting, masked arrays, and universal functions for complex mathematical operations.
- It integrates with C/C++ and Fortran for extending functionality in scientific computing.
- NumPy provides FFTs, linear algebra routines, and random sampling efficiently.

19. How does Pandas simplify time series analysis
- Pandas offers datetime indexing, resampling, and frequency conversion tools tailored for temporal data.
- It simplifies lag, rolling, and window operations for trend and seasonality analysis.
- Time zone handling and shifting functions make global time series analysis seamless.

20. What is the role of a pivot table in Pandas
- Pivot tables summarize data by reorganizing columns and rows, supporting aggregation they simplify comparison, trend analysis, and categorical grouping in tabular formats.
- Pivot tables are widely used for interactive reporting and multidimensional analysis.

21. Why is NumPy’s array slicing faster than Python’s list slicing
- NumPy slicing returns views instead of copies, reducing memory overhead.
- Arrays use contiguous memory blocks, enabling faster element access than lists. NumPy leverages C-level optimizations, providing efficient low-level slicing performance.

22. What are some common use cases for Seaborn
- Seaborn is used for creating heatmaps, violin plots, and regression visualizations.
- It simplifies visualizing categorical comparisons, statistical distributions, and pairwise relationships.
- Seaborn is frequently applied in exploratory analysis and reporting tasks.

```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row

import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = arr.sum(axis=1)
print(row_sums)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame

import pandas as pd
df = pd.DataFrame({"A": [10, 20, 30], "B": [5, 15, 25]})
print(df["A"].mean())
```

```python
# 3. Create a scatter plot using Matplotlib

import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.scatter(x, y)
plt.show()
```

```python
# 4. Calculate correlation matrix using Seaborn and visualize with heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({"A": [1,2,3,4], "B": [4,3,2,1], "C":[5,6,7,8]})
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
# 5. Generate a bar plot using Plotly

import plotly.express as px
fig = px.bar(x=["A", "B", "C"], y=[10, 20, 15])
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column

import pandas as pd
df = pd.DataFrame({"A": [1, 2, 3]})
df["B"] = df["A"] * 2
print(df)
```

```python
# 7. Perform element-wise multiplication of two NumPy arrays

import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a * b)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt
x = [1, 2, 3]
plt.plot(x, [1, 2, 3], label="Line1")
plt.plot(x, [3, 2, 1], label="Line2")
plt.legend()
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows greater than a threshold

import pandas as pd
df = pd.DataFrame({"A": [5, 10, 15, 20]})
print(df[df["A"] > 10])
```

```python
# 10. Create a histogram using Seaborn

import seaborn as sns
import matplotlib.pyplot as plt
data = [1,2,2,3,3,3,4,4,4,4]
sns.histplot(data, bins=4, kde=True)
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy

import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(np.dot(a, b))
```
